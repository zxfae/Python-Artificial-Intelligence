```python
import numpy as np
import matplotlib.pyplot as plt

class StandardCoffeeDecision:
    def __init__(self):
        self.name = "Standard Sigmoid: Décision basée sur l'envie"
    
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-x))
    
    def decide(self, desire_level):
        probability = self.sigmoid(desire_level)
        return probability, "acheter" if np.random.random() < probability else "ne pas acheter"
```

```python

class StretchedCoffeeDecision:
    def __init__(self):
        self.name = "Stretched Sigmoid: Décision basée sur le temps écoulé"
    
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-(0.5*x + 3)))
    
    def decide(self, hours_since_last_coffee):
        probability = self.sigmoid(hours_since_last_coffee)
        return probability, "acheter" if np.random.random() < probability else "ne pas acheter"
```

```python

class CompressedCoffeeDecision:
    def __init__(self):
        self.name = "Compressed Sigmoid: Décision basée sur le prix"
    
    def sigmoid(self, x):
        return 1 / (1 + np.exp(-(5*x + 11)))
    
    def decide(self, price):

        probability = self.sigmoid(-price)
        return probability, "acheter" if np.random.random() < probability else "ne pas acheter"
```

```python

def test_decisions():
    decisions = [StandardCoffeeDecision(), StretchedCoffeeDecision(), CompressedCoffeeDecision()]
    
    for decision in decisions:
        print(f"\nTest de {decision.name}")
        
        if isinstance(decision, StandardCoffeeDecision):
            inputs = [-4, 0, 4]
            input_name = "Niveau d'envie"
        elif isinstance(decision, StretchedCoffeeDecision):
            inputs = [1, 5, 9]
            input_name = "Heures depuis le dernier café"
        else:
            inputs = [1.5, 2.5, 3.5]
            input_name = "Prix du café (€)"
        
        for input_value in inputs:
            prob, choice = decision.decide(input_value)
            print(f"{input_name}: {input_value}, Probabilité: {prob:.2f}, Décision: {choice}")

test_decisions()
```

```python

x = np.linspace(-5, 5, 100)
plt.figure(figsize=(12, 4))

plt.subplot(131)
plt.plot(x, StandardCoffeeDecision().sigmoid(x))
plt.title("Standard Sigmoid")
plt.xlabel("Niveau d'envie")
plt.ylabel("Probabilité d'achat")

plt.subplot(132)
plt.plot(x, StretchedCoffeeDecision().sigmoid(x))
plt.title("Stretched Sigmoid")
plt.xlabel("Heures depuis le dernier café")
plt.ylabel("Probabilité d'achat")

plt.subplot(133)
plt.plot(x, CompressedCoffeeDecision().sigmoid(-x))
plt.title("Compressed Sigmoid")
plt.xlabel("Prix du café (€)")
plt.ylabel("Probabilité d'achat")

plt.tight_layout()
plt.show()
```
