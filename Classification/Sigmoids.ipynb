```python
import numpy as np
import matplotlib.pyplot as plt
```

#### 1 - Def Sigmoid

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid1(x):
    return 1 / (1 + np.exp(-(0.5*x + 3)))

def sigmoid2(x):
    return 1 / (1 + np.exp(-(5*x + 11)))
```

#### 2 - Plot Sigmoid
###### plot isn't my fav part

```python

# Create a range of x values
x = np.linspace(-20, 15, 1000)

# Create the plot
plt.figure(figsize=(10, 6))

# Plot the sigmoid functions
plt.plot(x, sigmoid(x), label='Standard Sigmoid', color='blue')
plt.plot(x, sigmoid1(x), label='Sigmoid1', color='orange')
plt.plot(x, sigmoid2(x), label='Sigmoid2', color='green')

# Add a horizontal line at y=0.5
plt.axhline(y=0.5, color='red', linestyle='-', label='p=0.5')

# Customize the plot
plt.title('Sigmoids')
plt.xlabel('x')
plt.ylabel('')  # Remove y-axis label
plt.legend()

# Set axis limits
plt.xlim(-20, 15)
plt.ylim(0, 1.1)

# Remove top and right spines
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# Show the plot
plt.show()
```
