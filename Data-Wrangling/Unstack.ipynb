```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Create business dates
business_dates = pd.bdate_range('2021-01-01', '2021-12-31')

# Generate tickers
tickers = ['AAPL', 'FB', 'GE', 'AMZN', 'DAI']

# Create MultiIndex
index = pd.MultiIndex.from_product([business_dates, tickers], names=['Date', 'Ticker'])

# Create DataFrame
market_data = pd.DataFrame(index=index,
                           data=np.random.randn(len(index), 1),
                           columns=['Prediction'])

# Unstack the DataFrame
unstacked_data = market_data.unstack(level='Ticker')

# Display the first 3 rows
print(unstacked_data.head(3).to_markdown())
```

```python

ax = unstacked_data.plot(figsize=(12, 6), title='Daily Predictions for Companies')

plt.xlabel('Date')
plt.ylabel('Prediction')
plt.legend(title='Company', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

plt.show()
```
