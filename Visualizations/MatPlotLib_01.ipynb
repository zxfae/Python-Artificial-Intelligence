```python
import matplotlib.pyplot as plt
```

```python
# Datas
x = [1, 4, 5, 6, 7]
y = [2, 6, 3, 6, 3]
```

```python
# Plot style
plt.plot(x, y, color='red', linestyle='-.', linewidth=2)
plt.scatter(x, y, color='blue', s=12**2)
plt.title('Display marker', fontsize=10)
plt.xlabel('x - axis', fontsize=12)
plt.ylabel('y - axis', fontsize=12)
plt.xlim(1, 8)
plt.ylim(1, 8)
```
