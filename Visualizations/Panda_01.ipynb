```python
import pandas as pd
```

```python
# Datas
df = pd.DataFrame({
        'name':['christopher','marion','maria','mia','clement','randy','remi'],
        'age':[70,30,22,19,45,33,20],
},
        index=['christopher','marion','maria','mia','clement','randy','remi']
        )
```

```python
# plot
plot = df.plot(kind='bar', title="Age per name", legend="name")
```
