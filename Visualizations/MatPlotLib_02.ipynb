```python
import matplotlib.pyplot as plt
```

```python
# Data
left_data = [5, 7, 11, 13, 17]
right_data = [0.1, 0.2, 0.4, 0.8, -1.6]
x_axis = [0.0, 1.0, 2.0, 3.0, 4.0]
```

```python

# Create the figure and the first axis
fig, ax1 = plt.subplots(figsize=(10, 6))
# First line
color1 = 'black'
ax1.set_ylabel('Big', color=color1)
ax1.plot(x_axis, left_data, color=color1)

# Create the twin
ax2 = ax1.twinx()

# Second line, second axis
color2 = 'red'
ax2.set_ylabel('Small', color='black')
ax2.plot(x_axis, right_data, color=color2)

plt.title('Twin axes', fontsize=16)

fig.tight_layout()
```
