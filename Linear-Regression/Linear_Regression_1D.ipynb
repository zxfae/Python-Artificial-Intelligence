```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_regression
from sklearn.linear_model import LinearRegression
```

#### Datas

```python
# Create datas test
# Testing with differents noises and complexity
X, y, coef = make_regression(n_samples=100, # 100 Datas IND
                             n_features=1, # 1 Datas DEP
                             n_informative=1,
                             noise=50, # Disperses Datas
                             coef=True,
                             random_state=0, # Seed
                             bias=100.0)
```

#### 1 - Plot the datas

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', alpha=0.7)
plt.title('X (1 dimension) and y')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

#### 2 - Fit a LinearRegression from Scikit-learn on the generated data and give the equation of the fitted line. The expected output is: y = coef * x + intercept

```python
model = LinearRegression().fit(X,y)
print("Coef",model.coef_)
print("Intercept",model.intercept_)
print("Score",model.score(X,y))
print(f"y = {model.coef_[0]} * x + {model.intercept_}")
```

#### 3 - Add the fitted line to the plot. The plot should look like this:

```python
plt.figure(figsize=(10, 6))
plt.scatter(X, y, color='blue', alpha=0.7)
plt.plot(X, model.predict(X), color='red', linewidth=2)
plt.title('X (1 dimension) and y')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

#### 4 - Predict on X.

```python
ten_first = X[:10]
print(f"Prediction : {model.predict(ten_first)}")
```

#### 5 - Create a function that computes the Mean Squared Error (MSE) and compute the MSE on the data set. The MSE is frequently used as well as other regression metrics that will be studied later this week.

```python
# Define the MSE function
def compute_mse(y_true, y_pred):
    #y_true = true value, y_pred = value predict
    return np.mean((y_true - y_pred)**2)

# Compute MSE on the entire dataset
y_pred_all = model.predict(X)
mse = compute_mse(y, y_pred_all)

print(f"\nMean Squared Error on the dataset: {mse}")
```
